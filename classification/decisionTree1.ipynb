```python
# loading module
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree # for creating Tree in visual form 
```

```python
# loading data
iris=load_iris()
iris.target_names # labels name 
iris.feature_names # features name
features=iris.data # actual data of features 
labels=iris.target # actual data of labels 
```

```python
# only setosa 
setosa=features[:50]
plt.xlabel('sepal_petal_L')
plt.ylabel('sepaL_petal_W')
plt.title('setosa data')
plt.scatter(setosa[:,0],setosa[:,1],c='red')
plt.scatter(setosa[:,2],setosa[:,3],c="green",marker="*")
plt.show()
```

```python
# lets split data into training & testing 
trainX,testX,TrainY,TestY=train_test_split(features,labels,test_size=0.2)
```

```python
# calling model 
model=DecisionTreeClassifier() # calling classifier 
# training model 
desc_trained=model.fit(trainX,TrainY)
```

```python
# predicting 
predicted_output=desc_trained.predict(testX)
predicted_output
```

```python
# actual output 
TestY
```

```python
# finding accuracy score 
score=accuracy_score(TestY,predicted_output)
print("accuracy of DecisionTree is ",score)
```

```python
# lets plot tree structure 
plt.figure(figsize=(12,8)) # defining image size
plot_tree(desc_trained ,feature_names=iris.feature_names, class_names=iris.target_names.tolist(),filled=True)
plt.title('Decision Tree')
plt.show()
```
