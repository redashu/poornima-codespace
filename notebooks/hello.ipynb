```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sb 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
# loading dataset from http uRL 
df=pd.read_csv('https://raw.githubusercontent.com/redashu/Datasets/master/advertising.csv')
```

```python
# lets plot scatter plot using seaborn 
sb.pairplot(df,y_vars=['Sales'],x_vars=['TV','Newspaper','Radio'],kind='scatter')
plt.show()
```

```python
# creating dataset for training model 
features=df[['TV','Radio','Newspaper']]
#features1=df.iloc[:,:-1].values
labels=df['Sales']
# lets split data
train_ft,test_ft,train_lb,test_lb=train_test_split(features,labels,test_size=0.1,random_state=100)
# train_ft is 90% of features and test_ft is 10% of features
# train_lb is 90% of lables , test_lb is 10% of lables 
```

```python
# lets train data to the model
model=LinearRegression() # calling Linear regression maths formula 
# to train data into model 
trained_model=model.fit(train_ft,train_lb)
# printing intercept and coefficient 
coff=trained_model.coef_
intercept=trained_model.intercept_
```

```python
# lets predict
predicted_output=trained_model.predict(test_ft)
# predicted output 
print(predicted_output)
```

```python
# actual output 
#print(test_lb)
# lets plat scatter using matplot b/w actual and predicted output 
plt.xlabel('actual output')
plt.ylabel('Predicted Output')
plt.title('Advertise prediction')
plt.scatter(test_lb,predicted_output)
# lets draw a best fit line as per my dataset 
plt.plot([min(test_lb),max(test_lb)],[min(predicted_output),max(predicted_output)],color='red',linestyle="dotted")
plt.show()
```
